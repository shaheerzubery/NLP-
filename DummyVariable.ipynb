### Dummy Variable - Preprocessing Tech

## Import Libraries

```python
import pandas as pd
import numpy as np
```

## Data Import

we are using the same dataset which we used to demonstrate Onehotencoding method

```python
df = pd.read_excel('/content/One hot encoding data_set.xlsx')
df.head()
```

here in this data we have two categorical columns (COlor and state)

```python
colors = pd.get_dummies(df.Color)
```

```python
colors
```

So here we have all the colors

we do the same for state as well

```python
States =pd.get_dummies(df.State)
```

Now we will merge all the list together

```python
df_new = pd.concat([df,colors,States] , axis = 1)
```

```python
df_new
```

There is One issue with Dummy variables, it is good for less one for example if there are 4 column it is good for three and if it has 3 column it is good for 2 

So we can drop it the columns

```python
final = df_new.drop(['Color', 'Red'], axis = 'columns')
```

```python
final = final.drop(['State', 'CA'], axis = 'columns')
```
